```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


airline = pd.read_csv("./customer_airways_data.csv", encoding="iso-8859-1")
airline.drop_duplicates(inplace=True)
```

#### Pregunta de interes: Existe una relacion entre las preferencias tomadas por los clientes, tipo de viaje y el horario de partida ?

```python
airline = airline[['wants_extra_baggage', 'wants_preferred_seat','wants_in_flight_meals','trip_type','flight_hour']]
```

#### Relacion entre el horario de partida y preferencias

```python
sns.lineplot(data=airline,x="flight_hour",y='wants_extra_baggage',linestyle="solid",color="darkred")
sns.lineplot(data=airline,x="flight_hour",y='wants_in_flight_meals',linestyle="solid",color="darkblue")
sns.lineplot(data=airline,x="flight_hour",y="wants_preferred_seat",linestyle="solid",color="darkgreen")
plt.ylabel('preferred_seat       flight_meals       extra_baggage')
plt.title("relationship between departure time and preferences")
```

#### Observamos que la franja horaria donde tenemos mejor promedio en cuanto a eleccion de preferencias es:

  1. Entre 06 - 11 hs para equipaje extra
  2. Entre 09 - 11 js para comida preferida
  3. Entre 10 - 13 hs y 22 - 23 hs para asiento preferido

#### Hay relacion con el tipo de viaje para estas preferencias ?

```python
sns.pointplot(data=airline,x="trip_type",y='wants_extra_baggage',linestyle="solid",color="darkred")
sns.pointplot(data=airline,x="trip_type",y='wants_in_flight_meals',linestyle="solid",color="darkblue")
sns.pointplot(data=airline,x="trip_type",y="wants_preferred_seat",linestyle="solid",color="darkgreen")
plt.ylabel('preferred_seat       flight_meals       extra_baggage')
plt.title("relationship between departure time and preferences")
```

### Preferencias del sitio:

#### + No se observa una relacin significativa entre la preferencia del sitio y el tipo de viaje, por lo que podemos decir que la preferencia del sitio no parece ser un factor determinante en la eleccin del tipo de viaje.
#### + Comida preferida se solicita con menos frecuencia en viajes de OneWay, respecto a los otros dos tipos, manteniendo el promedio de solicitudes inferior al 50%.
#### + En el equipaje extra observamos un notable aumento en la solicitud en viajes por tramos (CircleTrip). El promedio de solicitudes de equipaje extra supera el 60% en todos los tipos de viaje.




#### Seria interesante analizar si  disminucion del equipaje extra se debe a que el numero de solicitante de reserva es mayor:


```python
airline = pd.read_csv("./customer_airways_data.csv", encoding="iso-8859-1")
airline.drop_duplicates(inplace=True)

airline = airline[['wants_extra_baggage', 'num_passengers']]

sns.lineplot(data=airline,x="num_passengers",y='wants_extra_baggage',linestyle="solid",color="darkred")
```

#### Se muestra como a medida que aumenta el numero de pasajeros, es mayor (en promedio) la adquisicion del equipaje extra, con un caso particular para ocho personas en el cual el promedio baja considerablemente
