#### Busco relacionar comentarios de purchase lead con el purchase_lead del airline

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

def menciona (comment):
  # Lista de palabras clave relacionadas con "purchase_lead"
  keywords = ["purchase_lead", "purchase lead", "lead", "sales lead", "buy lead"]
  # Tokenizamos el texto del comentario
  words = word_tokenize(comment)
  # Verificar si alguna palabra clave está presente en el comentario
  mentioned_purchase_lead = any(keyword.lower() in words for keyword in keywords)

  if mentioned_purchase_lead:
    return 1
  
  return 0

reviews_filter_purchase_lead = reviews
reviews_filter_purchase_lead["purchase_lead"] = reviews.reviews.map(menciona)


reviews_filter_purchase_lead.purchase_lead.value_counts()

reviews_filter_purchase_lead.loc[reviews_filter_purchase_lead.purchase_lead == 1,:]

```
